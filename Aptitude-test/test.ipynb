```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('data/aptitude-test-group01.csv')
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
types = dataset['Type']
```

```python
types
```

```python
my_list = [0]*20 + [1]*20
np.random.shuffle(my_list)
```

```python
my_list = pd.DataFrame(my_list,columns=['Student answers'])
```

```python
student_correct_answer = pd.merge(types,my_list,right_index=True,left_index=True)
```

```python
student_correct_answer
```

```python
total_correct_answers = student_correct_answer['Student answers'].values.sum()
```

```python
print(f'Accuracy of student : {total_correct_answers/student_correct_answer.shape[0]*100}%')
```

```python
grouped_data = student_correct_answer.groupby('Type').mean()
```

```python
grouped_data
```

```python
labels = student_correct_answer['Type'].unique()
```

```python
plt.pie(grouped_data['Student answers'].values, labels=labels)
plt.show()
```
